```python
# prompt: series de tiempo con mas detalle

import pandas as pd
import matplotlib.pyplot as plt

# Create a dataframe with timestamps and values
data = {'timestamp': pd.to_datetime(['2020-01-01', '2020-01-02', '2020-01-03', '2020-01-04', '2020-01-05']),
       'value': [10, 15, 20, 18, 22]}
df = pd.DataFrame(data)

# Set the timestamp as the index
df = df.set_index('timestamp')

# Plot the data
df['value'].plot(figsize=(12, 6))
plt.xlabel('t(h)')
plt.ylabel('NOX (PPB)')
plt.title('NOX en la Merced 2006')
plt.show()
```
